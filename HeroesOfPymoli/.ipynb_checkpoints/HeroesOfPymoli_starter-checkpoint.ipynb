```python
# Dependencies and Setup
import pandas as pd
import os 

#file = os.path.join("", "Resources", "purchase_data.csv")


# # File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# # Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#rename dataframe
df = purchase_data
#establish number of players by counting unique screen names only
total_number_of_players = len(df["SN"].unique())
total_number_of_players
#build table
Total_Players_Table = pd.DataFrame([{"Total Players": total_number_of_players}])
Total_Players_Table

```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#establish number of unique items
num_of_unique_items = len(df["Item ID"].unique())

#establish average purchase price
avg_purchase_price = (df["Price"].mean())
#establish the total number of purchases
total_num_of_purchases = (df["Purchase ID"].count())
#establish the total revenue
total_revenue = (df["Price"].sum())
#build the summary table
summary_df = pd.DataFrame([{"Number of Unique Items": num_of_unique_items, 
                            "Average Price": avg_purchase_price, 
                            "Number of Purchases": total_num_of_purchases, 
                            "Total Revenue": total_revenue
                           }])
#format the summary table
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)
#call the summary table
summary_df

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#establish new dataframe minus duplicate screen names
gender_total = df.drop_duplicates("SN")
gender_total  
#establish new column that counts gender total
total_count = gender_total["Gender"].value_counts()
total_count
#determine the percentage of players
percentage_of_players = total_count/total_number_of_players*100

#build table
gender_demographic_df = pd.DataFrame({"Total Count": total_count, 
                                      "Percentage of Players": percentage_of_players
                                     })
#format table
gender_demographic_df["Percentage of Players"] = gender_demographic_df["Percentage of Players"].map("{0:.2f}%".format)
#call table
gender_demographic_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#establish individual purchase count totals for gender
total_count = gender_total["Gender"].value_counts()
                 
#establish average purchase price                  
avg_purchase_price = gender_total.groupby("Gender")["Price"].mean()

#establish total purchase value
total_purchase_value = total_count*avg_purchase_price
#establish average total purchase per person
avg_total_purchase = df.groupby("Gender").sum()["Price"]/total_count
#build table
gender_purchasing_analysis = pd.DataFrame({"Purchase Count": total_count, 
                                           "Average Purchase Price": avg_purchase_price, 
                                           "Total Purchase Value": total_purchase_value, 
                                           "Avg Total Purchase per Person": avg_total_purchase
                                          })
#formt table
gender_purchasing_analysis["Average Purchase Price"] = gender_purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
gender_purchasing_analysis["Total Purchase Value"] = gender_purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
gender_purchasing_analysis["Avg Total Purchase per Person"] = gender_purchasing_analysis["Avg Total Purchase per Person"].map("${:,.2f}".format)
#call table
gender_purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#determine total purchase value
total = df.groupby(["SN"]).sum()["Price"]
#determine average purchase price
average = df.groupby(["SN"]).mean()["Price"]
#determine purchase count
count = df.groupby(["SN"]).count()["Price"]
#build table
Top_Spender_df = pd.DataFrame({"Total Purchase Value": total,
                              "Average Purchase Price": average,
                              "Purchase Count": count
                              })
#sort table by highest "Total Purchase Value"
Top_Spender_df = Top_Spender_df.sort_values("Total Purchase Value", ascending = False)
#format table
Top_Spender_df["Total Purchase Value"] = Top_Spender_df["Total Purchase Value"].map("${:,.2f}".format)
Top_Spender_df["Average Purchase Price"] = Top_Spender_df["Average Purchase Price"].map("${:,.2f}".format)
#reorganize table
Final_Top_Spender_df = Top_Spender_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]] 
#call table
Final_Top_Spender_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#establish new dataframe 
item_data = df.loc[:, ["Item ID", "Item Name", "Price"]]
#establish item price
average_item_purchase = item_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
#establish total purchase value
total_item_purchase = item_data.groupby(["Item ID", "Item Name"]).sum()["Price"]
#establish purchase count
total_item_count = item_data.groupby(["Item ID", "Item Name"]).count()["Price"]

#build table
item_data_df = pd.DataFrame({"Total Purchase Value": total_item_purchase,
                            "Item Price": average_item_purchase, 
                            "Purchase Count": total_item_count})
#sort table by highest purchase count
item_data_sorted = item_data_df.sort_values("Purchase Count", ascending = False)
#format table
item_data_sorted["Item Price"] = item_data_sorted["Item Price"].map("${:,.2f}".format)
item_data_sorted["Total Purchase Value"] = item_data_sorted["Total Purchase Value"].map("${:,.2f}".format)
#call table
item_data_sorted


```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#use previous dataframe and sort on Total Purchase Value rather than Purchase Count
profitable_items_sorted = item_data_df.sort_values("Total Purchase Value", ascending = False)
#format table
profitable_items_sorted["Item Price"] = profitable_items_sorted["Item Price"].map("${:,.2f}".format)
profitable_items_sorted["Total Purchase Value"] = profitable_items_sorted["Total Purchase Value"].map("${:,.2f}".format)
#call table
profitable_items_sorted
```
